```python
x = linspace(0, 1, 1000)
astarts = geomspace(0.05, 10, 8)
bstarts = geomspace(0.05, 10, 8)
for a, b in sorted(product(astarts, bstarts), key=lambda x: x[0] + x[1]):
    y = beta.pdf(x, a, b)
    if y.max() != np.inf:
        plt.plot(x, y / y.max())

```

```python
def func(params, x1, p1, x2, p2, x3, p3):
    a = params[0]
    b = params[1]
    loc = params[2]
    scale = params[3]
    dist = beta(a, b, loc, scale)
    cdf0 = dist.cdf(0)
    cdf1 = dist.cdf(x1)
    cdf2 = dist.cdf(x2)
    cdf3 = dist.cdf(x3)
    result = cdf0 ** 2 +  ((cdf1 - p1) ** 2 + (cdf2 - p2) ** 2) + (cdf3 - p3) ** 2
    return result

def find_beta(x1, p1, x2, p2, x3, p3):
    assert x1 < x2 < x3
    assert p1 < p2 < p3
    astarts = geomspace(0.05, 10, 8)
    bstarts = geomspace(0.05, 10, 8)
    locstart = x1
    scalestart = x3 - x1
    
    minresult = None
    minresultfun = float('inf')
    for astart, bstart in product(astarts, bstarts):

        result = minimize(func, [astart, bstart, locstart, scalestart], args=(x1, p1, x2, p2, x3, p3), method='Nelder-Mead', options={'adaptive': True})
        if result.fun < minresultfun:
            print(f"{astart}, {bstart}")
            print(result.fun)
            minresultfun = result.fun
            minresult = result
    print(result)
    return beta(*list(result.x))
        
    
    

# result = minimize(func, [2, 5, 3, 5], args=(3, 0.1, 5, 0.5, 8, 0.9), method='Nelder-Mead', options={'adaptive': True})
rdist = find_beta(3, 0.1, 5, 0.5, 8, 0.9)


rdist.cdf([3, 5, 8])
```
